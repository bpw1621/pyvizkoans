```python
from collections import deque
import calendar

from matplotlib import rc
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# rc('text', usetex=True)
# rc('text.latex', preamble=r'\usepackage{color}')
```

```python
VACATION_STEPS_COLOR = '#446cb5'
# AVERAGE_STEPS_COLOR = '#ababab'
AVERAGE_STEPS_COLOR = '#737373'
```

```python
xaxis_labels = deque(day_name.upper()[:3] for day_name in calendar.day_name)
xaxis_labels.rotate(2)
```

```python
blue_line = [15_800, 12_000, 13_000, 12_800, 14_000, 9_600, 11_000]
gray_line = [y - 5_000 + 1_000 * np.random.randn() for y in blue_line]
```

```python
fig, ax = plt.subplots(figsize=(12, 6), tight_layout=True)

ax.plot(blue_line, color=VACATION_STEPS_COLOR, lw=2.5)
ax.scatter([0, len(blue_line) - 2], [blue_line[0], blue_line[-2]], s=[100, 100])
ax.plot(gray_line, color=AVERAGE_STEPS_COLOR, lw=2.5)

sns.despine(ax=ax, left=True)
ax.yaxis.set_visible(False)

ax.xaxis.set_ticks(range(7))
ax.xaxis.set_ticklabels(xaxis_labels)
ax.set_xlabel('DAY OF THE WEEK', loc='left')
ax.set_xlim((-0.05, 6))

ax.set_ylim((0, 20_000))

# ax.set_title(r'This is {\color{red} highlighted}, and this is not.')
ax.set_title(r'My steps on our road trip were 1.5x higher than my average', loc='left')

ax.annotate('VACATION STEPS', xy=(6, blue_line[-1]), textcoords='offset points', xytext=(8, -3), color=VACATION_STEPS_COLOR)
ax.annotate('AVERAGE', (6, gray_line[-1]), textcoords='offset points', xytext=(8, -3), color=AVERAGE_STEPS_COLOR)
```
